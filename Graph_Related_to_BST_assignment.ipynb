```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for execution time and memory usage
N_values = np.array([10, 100, 1000, 5000])
recursive_time_complete = np.array([0.001018, 0.009153, 0.106255, 0.558826])
iterative_time_complete = np.array([0.000246, 0.001877, 0.012925, 0.049470])
recursive_time_skewed = np.array([0.000861, 0.013819, 0.561993, 11.559169])
iterative_time_skewed = np.array([0.000300, 0.001512, 0.008699, 0.034079])

# Memory usage (hypothetical in KB)
recursive_memory_complete = np.array([416, 1744, 16144, 80144])
iterative_memory_complete = np.array([192, 928, 8896, 41920])
recursive_memory_skewed = np.array([192, 1632, 16032, 80032])
iterative_memory_skewed = np.array([192, 1312, 12480, 62336])

# Create time complexity plot
plt.figure(figsize=(10, 5))
plt.plot(N_values, recursive_time_complete, label="Recursive - Complete", marker='o')
plt.plot(N_values, iterative_time_complete, label="Iterative - Complete", marker='s')
plt.plot(N_values, recursive_time_skewed, label="Recursive - Skewed", marker='o', linestyle='dashed')
plt.plot(N_values, iterative_time_skewed, label="Iterative - Skewed", marker='s', linestyle='dashed')

plt.xlabel("Number of Nodes (N)")
plt.ylabel("Execution Time (seconds)")
plt.title("Execution Time Comparison")
plt.legend()
plt.grid(True)
plt.show()

# Create memory usage plot
plt.figure(figsize=(10, 5))
plt.plot(N_values, recursive_memory_complete, label="Recursive - Complete", marker='o')
plt.plot(N_values, iterative_memory_complete, label="Iterative - Complete", marker='s')
plt.plot(N_values, recursive_memory_skewed, label="Recursive - Skewed", marker='o', linestyle='dashed')
plt.plot(N_values, iterative_memory_skewed, label="Iterative - Skewed", marker='s', linestyle='dashed')

plt.xlabel("Number of Nodes (N)")
plt.ylabel("Memory Usage (Bytes)")
plt.title("Memory Usage Comparison")
plt.legend()
plt.grid(True)
plt.show()
```
